```python
# Import dependencies
import pandas as pd
```

```python
# Import our first dataset (USAF data from new website, aviation archaeology)
file_path = "/Users/clarapotts/Desktop/Capstone_Project copy/Capstone Project/Datasets/Raw/Vietnam_USAF_Losses_Raw_Data.csv"
usaf_df = pd.read_csv(file_path)
usaf_df.head()
```

```python
# Want to split single column into multiple using ':' but a few field names have no ':' after them so adding here.
# Also cleaning up some other field names so split doesn't cause problems.
to_replace = ["Call Sign", "Mission Type", "Where Hit", "Pilot Egress", "Pilot Status", "Pilot Condition",
             "Pilot Recovered", "Co-Pilot/Nav Rank", "Co-pilot/Nav", "Co-Pilot/Nav Egress", "Co-Pilot/Nav Status",
             "Co-Pilot/Nav Condition", "Co-Pilot/Nav Recovered"]
replace_with = ["Call_Sign:", "Mission_Type:", "Where_Hit:", "PILOT_EGRESS", "PILOT_STATUS", "PILOT_CONDITION",
               "PILOT_RECOVERED", "CP_NAV_RANK", "CP_NAV", "CP_EGRESS", "CP_STATUS",
               "CP_CONDITION", "CP_RECOVERED"]


usaf_df = usaf_df.replace(to_replace, replace_with, regex=True)
```

```python
# Now to make all field names disappear.
to_replace = ["Crash Date", "Crash Time", "Aircraft Type", "Aircraft S/N", "Base", "Wing", "Squadron", "Call_Sign",
              "Operations Code", "Mission_Type", "Weapon", "Target Objective", "Ceiling/Vis", "JCS NR", "Route Pack",
              "Maneuver", "Pass", "Angle", "Altitude", "Airspeed", "Mission Phase", "Where_Hit", "Fire Observed",
              "Hit Country", "Hit Longitude", "Hit Latitude", "Loss Country", "Loss Longitude", "Loss Latitude",
              "Defense Type", "Pilot Hit", "Pilot Rank", "Pilot", "PILOT_EGRESS", "PILOT_STATUS", "PILOT_CONDITION",
              "PILOT_RECOVERED", "CP_NAV_RANK", "CP_NAV", "CP_EGRESS", "CP_STATUS",
              "CP_CONDITION", "CP_RECOVERED"]

replace_with = ""
              
usaf_df = usaf_df.replace(to_replace, replace_with, regex=True)
```

```python
# Breaking the dataframe into different columns
usaf_df = usaf_df['All_Info'].str.split(':', expand=True)

usaf_df.head()
```

```python
# Drop column '0' since no data contained here.
usaf_df.drop(columns={0}, inplace=True)
usaf_df.head()
```

```python
# Renaming the columns columns.
usaf_df.columns=["Crash_Date", "Crash_Time", "Aircraft_Type", "Aircraft_SN", "Base", "Wing",
                                     "Squadron", "Call_Sign", "Operations_Code", "Mission_Type", "Weapon", "Target_Objective",
                                     "Ceiling_Vis", "JCS_NR", "Route_Pack", "Maneuver", "Pass", "Angle", "Altitude",
                                     "Airspeed", "Mission_Phase", "Where_Hit", "Fire_Observed", "Hit_Country",
                                     "Hit_Longitude", "Hit_Latitude", "Loss_Country", "Loss_Longitude", "Loss_Latitude",
                                     "Defense_Type", "Pilot_Hit", "Pilot_Rank", "Pilot", "Pilot_Egress", "Pilot_Status",
                                     "Pilot_Condition", "Pilot_Recovered", "Co_Pilot_Nav_Rank", "Co_Pilot_Nav",
                                     "Co_Pilot_Nav_Egress", "Co_Pilot_Nav_Status", "Co_Pilot_Nav_Condition",
                                     "Co_Pilot_Recovered"]

usaf_df.head()
```

```python
# Checking for any nulls
usaf_df.isnull().any()
```

```python
# Looking for empty string rows (rows with no data but not picked up as nulls) so we can delete.
usaf_df.loc[usaf_df["Crash_Date"] == ''].count()
```

```python
# New data frame will just be rows with data
usaf_df = usaf_df.loc[usaf_df['Crash_Date'] != '']
usaf_df
```

```python
# Removing '0' from the Aircraft_SN column
usaf_df = usaf_df[usaf_df["Aircraft_SN"] != "0"]
usaf_df.head(20)
```

```python
# Convert aircraft serial number to integer
usaf_df = usaf_df.astype({"Aircraft_SN" : 'int'})
```

```python
# Check that conversion worked
aircraft_sn_datatype = usaf_df.dtypes["Aircraft_SN"]
print(f"Aircraft serial numbers are now data type: '{aircraft_sn_datatype}'")
```

```python
# Convert columns 'Pass', 'Angle', 'Altitude', 'Airspeed' to integers
usaf_df = usaf_df.astype({"Pass":'int', "Angle":'int', "Altitude":'int', "Airspeed":'int'})
```

```python
# Convert Crash_Date to datetime data type
usaf_df["Crash_Date"] = pd.to_datetime(usaf_df["Crash_Date"])
```

```python
usaf_df.head()
```

```python
# Correct Crash_Date years
usaf_df["Crash_Date"] = usaf_df["Crash_Date"] - pd.DateOffset(years=100)
```

```python
# Confirm correction
usaf_df.head()
```

```python
# Check to ensure conversions worked
print(usaf_df.dtypes)
```

```python
# Exporting to csv
usaf_df.to_csv('Vietnam_USAF_Losses_Cleaned_Data.csv', index=False)
```

```python
# Import our second dataset (from Vietnamairlosses.com).
## The original website I found but detailed info in paragraph form so just scraped the summary page to use for potential use
### as a second table for info like base name, etc.
file_path = "Resources/Raw/Vietnam_All_Fixedwing_Losses_Raw_Data.csv"
fixedwing_df = pd.read_csv(file_path)
fixedwing_df.head()
```

```python
# Cleaning up serial number so it matches format of the aviation archaeology website (no # or - ).
fixedwing_df["Aircraft_SN"] = fixedwing_df["Aircraft_SN"].str.replace("Ser","")\
    .str.replace("#","")\
    .str.replace("-","")\
    
fixedwing_df.head()
```

```python
# Export this other dataframe to its own csv
fixedwing_df.to_csv('Vietnam_All_Fixedwing_Losses_Data.csv', index=False)
```
